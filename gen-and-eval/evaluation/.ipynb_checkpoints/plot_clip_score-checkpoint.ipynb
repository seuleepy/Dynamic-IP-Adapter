Load clip score and plot

```python
import json
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python
# load json file

image_score_json = "./sd_v15/clip_image_score_vitl14.json"
text_score_json = "./sd_v15/clip_text_score_vitl14.json"
image_score_ti_json = "./sd_v15/clip_image_score_ti_vitl14.json"
text_score_ti_json = "./sd_v15/clip_text_score_ti_vitl14.json"

with open(image_score_json, 'r') as f:
    image_score = json.load(f)
with open(text_score_json, 'r') as f:
    text_score = json.load(f)

with open(image_score_ti_json, 'r') as f:
    image_score_ti = json.load(f)
with open(text_score_ti_json, 'r') as f:
    text_score_ti = json.load(f)
```

각 모델별로 image score, text score

```python
# summation per model

MODEL_LIST = [
        "constant_05",
        "constant_10",
        "linear_05",
        "linear_10",
        # "linear",
        # "relu_100",
        # "relu_500",
        # "relu_900",
        # "sigmoid",
        # "sine",
    ]
model_image_scores = defaultdict(float)
model_text_scores = defaultdict(float)

for key, score in image_score.items():
    for model in MODEL_LIST:
        if key.startswith(model):
            model_image_scores[model] += score
for key, score in image_score_ti.items():
    for model in MODEL_LIST:
        if key.startswith(model):
            model_image_scores[model] += score

for key, score in text_score.items():
    for model in MODEL_LIST:
        if key.startswith(model):
            model_text_scores[model] += score
for key, score in text_score_ti.items():
    for model in MODEL_LIST:
        if key.startswith(model):
            model_text_scores[model] += score

for key in model_image_scores:
    model_image_scores[key] /= 30
    model_text_scores[key] /= 30
```

```python
with open("./sd_v15/model_image_scores_ti.json", "w") as f:
    json.dump(model_image_scores, f, indent=4)
with open("./sd_v15/model_text_scores_ti.json", "w") as f:
    json.dump(model_text_scores, f, indent=4)
```

```python
models = list(model_image_scores.keys())
image_scores = list(model_image_scores.values())
plt.style.use(plt.style.available[10])
# plt.figure(figsize=(10, 6))
plt.plot(models, image_scores, marker='o', linestyle='-')
plt.xlabel('Model')
plt.ylabel('Score')
# plt.legend()
plt.title("CLIP Image Score")
plt.tight_layout()
plt.show()
```

```python
models = list(model_text_scores.keys())
text_scores = list(model_text_scores.values())
plt.style.use(plt.style.available[10])
# plt.figure(figsize=(10, 6))
plt.plot(models, text_scores, marker='o', linestyle='-')
plt.xlabel('Model')
plt.ylabel('Score')
# plt.legend()
plt.title("CLIP Text Score")
plt.tight_layout()
plt.show()
```

subject, model에 대해 image score, text score

```python
# comparison per target

from constant import TARGET2CLASS_MAPPING

MODEL_LIST = [
        "constant_05",
        "constant_10",
        "linear",
        "relu_100",
        "relu_500",
        "relu_900",
        "sigmoid",
        "sine",
    ]

target_image_dict = {}
target_text_dict = {}

for target in TARGET2CLASS_MAPPING:
    target_dict = {}
    for key, score in image_score.items():
        if key.endswith(target):
            model = key.removesuffix(f"_{target}")
            target_dict[model] = score
    target_image_dict[target] = target_dict
    with open(f"./realistic_vision_v40/{target}_image_scores.json", "w") as f:
        json.dump(target_dict, f, indent=4)
    
for target in TARGET2CLASS_MAPPING:
    target_dict = {}
    for key, score in text_score.items():
        if key.endswith(target):
            model = key.removesuffix(f"_{target}")
            target_dict[model] = score
    target_text_dict[target] = target_dict
    with open(f"./realistic_vision_v40/{target}_text_scores.json", "w") as f:
        json.dump(target_dict, f, indent=4)
```

```python
plt.figure(figsize=(10, 6))
for subject, scores in target_image_dict.items():
    model_scores = [scores[model] for model in MODEL_LIST]
    plt.plot(MODEL_LIST, model_scores, marker='o', label=subject)

plt.xlabel("Model")
plt.ylabel("Score")
plt.title("CLIP Image Score")
plt.legend()
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
for subject, scores in target_text_dict.items():
    model_scores = [scores[model] for model in MODEL_LIST]
    plt.plot(MODEL_LIST, model_scores, marker='o', label=subject)

plt.xlabel("Model")
plt.ylabel("Score")
plt.title("CLIP Text Score")
plt.legend()
plt.tight_layout()

plt.show()
```

subject (live vs object) 비교

```python
from constant import CLASS2OBJECT_MAPPING

live_image_dict = defaultdict(float)
object_image_dict = defaultdict(float)
live_text_dict = defaultdict(float)
object_text_dict = defaultdict(float)

live_len = 0
object_len = 0
for target, class_type in TARGET2CLASS_MAPPING.items():
    if CLASS2OBJECT_MAPPING[class_type]:
        image_dict = object_image_dict
        text_dict = object_text_dict
        object_len += 1
    else:
        image_dict = live_image_dict
        text_dict = live_text_dict
        live_len += 1
    for model in MODEL_LIST:
        image_dict[model] += target_image_dict[target][model]
        text_dict[model] += target_text_dict[target][model]

for model in MODEL_LIST:
    live_image_dict[model] /= live_len
    object_image_dict[model] /= object_len
    live_text_dict[model] /= live_len
    object_text_dict[model] /= object_len
```

```python
plt.figure(figsize=(10, 6))
live_image_scores = [live_image_dict[model] for model in MODEL_LIST]
object_image_scores = [object_image_dict[model] for model in MODEL_LIST]
plt.plot(MODEL_LIST, live_image_scores, marker='o', label="live")
plt.plot(MODEL_LIST, object_image_scores, marker='o', label="object")
plt.xlabel("Model")
plt.ylabel("Score")
plt.title("CLIP Image Score")
plt.legend()
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
live_image_scores = [live_text_dict[model] for model in MODEL_LIST]
object_image_scores = [object_text_dict[model] for model in MODEL_LIST]
plt.plot(MODEL_LIST, live_image_scores, marker='o', label="live")
plt.plot(MODEL_LIST, object_image_scores, marker='o', label="object")
plt.xlabel("Model")
plt.ylabel("Score")
plt.title("CLIP Text Score")
plt.legend()
plt.tight_layout()

plt.show()
```
